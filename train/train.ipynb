```python
import os
import shutil
from sklearn.model_selection import train_test_split
import pandas as pd
import keras
import tensorflow as tf
```

```python
df = pd.DataFrame()
#df.columns = ['melspect_path', 'audio_len', 'language', 'language_family', 'gender', 'emotion']
path = 'D:/licencjat/data/melspectrograms_csv'

for df_name in os.listdir(path):
    in_df = pd.read_csv(f'{path}/{df_name}')
    print(f'{path}/{df_name}')
    df = df.append(in_df)
    df = df.sample(frac=1).reset_index(drop=True)
```

```python
#df = df.drop('Unnamed: 0', axis=1)
df = df.loc[df['emotion'].isin(['angry', 'happy', 'neutral', 'sad'])]
df_train = df.loc[df['language'].isin(['deDE', 'enCA', 'enGB', 'enUS', 'esMX', 'frCA', 'grGR'])]
df_test = df_train[:2000]
df_train = df_train[2000:]
df_train.head(10)
```

```python
train_out_path = 'D:/licencjat/data/grouped/train'

for idx, row in df_train.iterrows():
    path = row['melspect_path']
    label = row['emotion']
    filename = os.path.split(path)[1]
    #print(f'{out_path}/{label}')

    if not os.path.exists(f'{train_out_path}/{label}'):
        os.makedirs(f'{train_out_path}/{label}')
    shutil.copyfile(path, f'{train_out_path}/{label}/{filename}')
```

```python
image_size = (65, 128)
batch_size = 128

train_dataset = tf.keras.preprocessing.image_dataset_from_directory('D:/licencjat/data/grouped/train', validation_split=0.2, subset="training", 
                                                                    seed=14, image_size=image_size, batch_size=batch_size, label_mode='categorical')
val_dataset  = tf.keras.preprocessing.image_dataset_from_directory('D:/licencjat/data/grouped/train', validation_split=0.2, subset="validation", 
                                                                    seed=14, image_size=image_size, batch_size=batch_size, label_mode='categorical')
```

```python
model = keras.models.Sequential()

model.add(keras.layers.Rescaling(1./255))

model.add(keras.layers.Conv2D(32, (3, 3), activation='relu', input_shape=(65, 128, 3)))
model.add(keras.layers.MaxPooling2D(3, 3))
model.add(keras.layers.BatchNormalization())

model.add(keras.layers.Conv2D(64, (3, 3), activation='relu', padding='same'))
model.add(keras.layers.MaxPooling2D(3, 3))
model.add(keras.layers.BatchNormalization())
model.add(keras.layers.Dropout(0.2))

model.add(keras.layers.Conv2D(128, (3, 3), activation='relu', padding='same'))
model.add(keras.layers.MaxPooling2D(2, 2))
model.add(keras.layers.BatchNormalization())
model.add(keras.layers.Dropout(0.2))

#model.add(keras.layers.Conv2D(256, (3, 3), activation='relu', padding='same'))
#model.add(keras.layers.MaxPooling2D(2, 2))
#model.add(keras.layers.BatchNormalization())
#model.add(keras.layers.Dropout(0.2))

#model.add(keras.layers.Conv2D(128, (3, 3), activation='relu', padding='same'))
#model.add(keras.layers.MaxPooling2D(2, 2))
#model.add(keras.layers.BatchNormalization())

#model.add(keras.layers.Conv2D(64, (3, 3), activation='relu', padding='same'))
#model.add(keras.layers.MaxPooling2D(2, 2))
#model.add(keras.layers.BatchNormalization())

model.add(keras.layers.Flatten())
model.add(keras.layers.Dense(128, activation='relu'))
model.add(keras.layers.Dropout(0.2))

model.add(keras.layers.Dense(64, activation='relu'))
model.add(keras.layers.Dropout(0.2))

model.add(keras.layers.Dense(32, activation='relu'))
model.add(keras.layers.Dense(5, activation='softmax'))
```

```python
model.compile(optimizer='adam', loss="categorical_crossentropy", metrics=["accuracy"])
history = model.fit(train_dataset, epochs=100, batch_size=64, validation_data=val_dataset)
```

```python
fser_model = keras.models.Sequential()

fser_model.add(keras.layers.Rescaling(1./255))

# Conv block 1
fser_model.add(keras.layers.Conv2D(filters=8, kernel_size=(5, 5), input_shape=input_shape, activation='relu'))
fser_model.add(keras.layers.MaxPooling2D(2, 2))
fser_model.add(keras.layers.Dropout(0.2))

# Conv block 2
fser_model.add(keras.layers.Conv2D(filters=16, kernel_size=(5, 5), activation='relu'))
fser_model.add(keras.layers.MaxPooling2D(2, 2))
fser_model.add(keras.layers.Dropout(0.2))

# Conv block 3
fser_model.add(keras.layers.Conv2D(filters=100, kernel_size=(5, 5), activation='relu'))
fser_model.add(keras.layers.MaxPooling2D(2, 2))
fser_model.add(keras.layers.Dropout(0.2))
# Conv block 4
fser_model.add(keras.layers.Conv2D(filters=200, kernel_size=(5, 5), activation='relu'))
fser_model.add(keras.layers.MaxPooling2D(2, 2))
fser_model.add(keras.layers.Dropout(0.2))

fser_model.add(keras.layers.Flatten())
fser_model.add(keras.layers.Dense(17424, activation='relu'))
fser_model.add(keras.layers.Dense(1024, activation='relu'))
fser_model.add(keras.layers.Dense(500, activation='relu'))
fser_model.add(keras.layers.Dense(5, activation='softmax'))
```

```python
batch_size = 64
epochs = 400
learnign_rate = 0.0001
```

```python
optimizer = tf.keras.optimizers.SGD(learning_rate=learnign_rate)
fser_model.compile(optimizer=optimizer, loss='categorical_crossentropy', metrics=['acc'])
fser_history = fser_model.fit(train_dataset, epochs=epochs, batch_size=batch_size, validation_data=val_dataset)
```

```python
resnet_model = keras.models.Sequential()
resnet = keras.applications.resnet.ResNet50(include_top=False, input_shape=(65, 128, 3))
for layer in resnet.layers:
    layer.trainable = False
```

```python
resnet_model.add(resnet)

resnet_model.add(keras.layers.Flatten())

resnet_model.add(keras.layers.Dense(512, activation='relu'))
resnet_model.add(keras.layers.Dropout(0.2))

resnet_model.add(keras.layers.Dense(128, activation='relu'))
resnet_model.add(keras.layers.Dropout(0.2))

resnet_model.add(keras.layers.Dense(5, activation='softmax'))
```
